```python
import os
import requests
import pandas as pd
import numpy as np
```

```python
#leer
file18 = '2018Tmax.csv'
file19 = '2019Tmax.csv'
df18 = pd.read_csv(file18, sep=',', engine='python')
df19 = pd.read_csv(file19, sep=',', engine='python')
df18
```

```python
#ver si existe una categoria
'Tmax'in df18
```

```python
#mostrar una categoria
df19['Tmax']
```

```python
#ver que indices tiene una tabla
df19.keys()
```

```python
list(df19.items())
```

```python
#aplicacion
comparar = df19

comparar['aumento'] = df19['Tmax']/df18['Tmax']
comparar
```

```python
#aumento de temperatura maxima en un año
comparar['aumento'].mean()
```

```python
#poner una tabla despues de otra
pd.concat([df18, df19], join = 'outer', sort = False)
```

```python
#Asi podemos sumar Tmax de cada tabla 
pd.concat([df18,df19],sort = False).groupby('Tmax').median()
```

```python
#mostrar el maximo de una tabla
df19.max()
```

```python
df18.max()
```
