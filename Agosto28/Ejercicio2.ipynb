```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
from scipy.integrate import odeint

x_0 = [0,3]
t = np.linspace(0,20,240)

def Sistema(x,t,g,l,m):
    l = 1   
    g = 9.81 
    m = 1   
    x1 = x[0]
    x2 = x[1]
    dx1_dt = x2
    dx2_dt = -g*math.sin(x1)
    dx_dt = [dx1_dt,dx2_dt]
    return dx_dt

%time X = odeint(Sistema,x_0,t,args = (g,l,m))

plt.plot(t,X[:,0],'b-')
plt.show()
```
